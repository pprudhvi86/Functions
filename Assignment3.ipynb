Q1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
#Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.

def sum_even(l):
  sum=0
  for i in l:
    if i%2==0:
      sum+=i
  return sum

sum_even([1,2,3,2,3,2,4,5,6,8,10,12])
```

Q4.Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
#Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(n):
  if n <=1:
    return False
  for i in range (2,int(n**0.5)+1):
    if n%i==0:
      return False
  return True

num=int(input("Enter a number between 1 and 200:"))
if prime(num):
  print(num,"is a prime number")
else:
  print(num,"is not a prime number")
```

Q5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b
for i in fibonacci(10):
  print(i)
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
#. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_gen(x):
  for i in range(x):
    yield 2**i
count=int(input ("Enter the power of 2:"))
for i in power_gen(count):
  print(i)
```

Q7. Implement a generator function that reads a file line by line and yields each line as a string.



```python
# Implement a generator function that reads a file line by line and yields each line as a string.
with open ("file.txt","w") as text:
  text.write("Hello \nPrudhvi\n This is a test file .")
def file_line(filename):
  with open(filename,'r') as file:
    for line in file:
      yield line
for line in file_line("file.txt"):
  print(line)
```

Q9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


```python
#Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
c=[0,45,38,22,11]
fahrenheit=list(map(lambda x: (x*9/5)+32,c))
print(fahrenheit)
```

Q10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
#Create a Python program that uses `filter()` to remove all the vowels from a given string.

Text=input("Please enter a sentence : ")
Vowels="aeiouAEIOU"
no_vowels=list(filter(lambda x: x not in Vowels,Text))
print("".join(no_vowels))
```
